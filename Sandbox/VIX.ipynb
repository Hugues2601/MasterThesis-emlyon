```python
import yfinance as yf
import matplotlib.pyplot as plt

# Téléchargement des données du VIX
vix = yf.download("^VIX", start="2004-01-01", end="2025-01-01")

# Création du graphique
plt.figure(figsize=(10, 5))
plt.plot(vix.index, vix["Close"], color="black")
plt.xlabel("Year")
plt.ylabel("VIX Value")
plt.grid(True)
plt.tight_layout()
plt.show()
```
